# Phân tích IoU
## Giới thiệu
    IoU (Intersection over Union) là một chỉ số quan trọng trong bài toán Object Detection, 
được sử dụng để đánh giá độ chính xác của bounding box mà mô hình dự đoán.

IoU đo lường mức độ chồng lắp giữa bounding box dự đoán (Prediction) và bounding box thực tế (Ground Truth), 
được tính bằng tỷ lệ giữa diện tích giao nhau và diện tích hợp của hai vùng này.

Giá trị IoU nằm trong khoảng từ 0 đến 1:
- IoU = 1: hai bounding box trùng khớp hoàn toàn
- IoU ≥ 0.5: dự đoán được xem là chấp nhận được
- IoU < 0.5: dự đoán chưa chính xác

Chỉ số này giúp đánh giá khả năng định vị (localization) của mô hình, 
tức là mô hình có vẽ đúng vị trí và kích thước của đối tượng hay không.

```python
COLORS = [(255, 0, 0), (0, 255, 0), (0, 0, 255), (255, 255, 0), (255, 0, 255), (0, 255, 255)]
def draw_ground_truth(img_path, label_path):
    """_summary_ Hàm vẽ bounding box Ground Truth lên ảnh dựa trên file nhãn YOLO.

    Args:
        img_path (_type_): _description_ đường dẫn tới ảnh gốc
        label_path (_type_): _description_ đường dẫn tới file nhãn tương ứng với ảnh, định dạng YOLO (class_id x_center y_center width height)

    Returns:
        _type_: _description_  trả về ảnh đã được vẽ bounding box GT và dict đếm số lượng từng class trong GT
    """
    img = cv2.imread(img_path)
    if img is None: 
        return None, {}

    h, w, _ = img.shape
    class_counts = {}

    if os.path.exists(label_path):
        with open(label_path, 'r') as f:
            for line in f:
                parts = list(map(float, line.strip().split()))
                if len(parts) >= 5:
                    class_id = int(parts[0])
                    class_name = CLASS_NAMES[class_id]

                    # đếm số lượng
                    class_counts[class_name] = class_counts.get(class_name, 0) + 1

                    color = COLORS[class_id % len(COLORS)]
                    x_c, y_c, box_w, box_h = parts[1:5]

                    # convert YOLO → pixel
                    x1 = int((x_c - box_w / 2) * w)
                    y1 = int((y_c - box_h / 2) * h)
                    x2 = int((x_c + box_w / 2) * w)
                    y2 = int((y_c + box_h / 2) * h)

                    # vẽ bbox
                    cv2.rectangle(img, (x1, y1), (x2, y2), color, 2)

                    # label = class name
                    cv2.putText(img, f"{class_name}", (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

    return img, class_counts

def get_pred_class_counts(results):
    """_summary_ Hàm đếm số lượng từng class trong kết quả dự đoán của mô hình YOLOv11 sau khi chạy predict.

    Args:
        results (_type_): _description_ đầu vào là kết quả dự đoán của mô hình YOLOv11 sau khi chạy predict, 
        chứa thông tin về bounding box và class_id của các đối tượng được phát hiện.

    Returns:
        _type_: _description_ trả về dict đếm số lượng từng class trong kết quả dự đoán, có dạng {class_name: count, ...}
    """
    counts = {}

    boxes = results[0].boxes
    if boxes is None or len(boxes) == 0:
        return counts

    class_ids = boxes.cls.cpu().numpy().astype(int)
    CLASS_NAMES = ['car', 'cycle', 'bus', 'truck', 'van']
    for cls_id in class_ids:
        class_name = CLASS_NAMES[cls_id]
        counts[class_name] = counts.get(class_name, 0) + 1

    return counts

def format_counts(counts):
    """_summary_ Hàm định dạng dict đếm số lượng từng class thành chuỗi để hiển thị trong báo cáo tổng kết.

    Args:
        counts (_type_): _description_ đầu vào là dict đếm số lượng từng class, có dạng {class_name: count, ...}

    Returns:
        _type_: _description_   trả về chuỗi định dạng để hiển thị, ví dụ: "car: 3, cycle: 1, bus: 0, truck: 2, van: 1". Nếu dict rỗng sẽ trả về "None".
    """
    if len(counts) == 0:
        return "None"
    return ", ".join([f"{k}: {v}" for k, v in counts.items()])



def get_gt_boxes_with_class(label_path, w, h):
    """_summary_ Hàm load Ground Truth từ file nhãn và chuyển đổi sang định dạng [x1, y1, x2, y2] kèm theo class_id để tính IoU với dự đoán của mô hình

    Args:
        label_path (_type_): _description_ đường dẫn tới file nhãn
        w (_type_): _description_ độ rộng của ảnh (dùng để chuyển đổi tọa độ chuẩn hóa sang pixel)
        h (_type_): _description_ độ cao của ảnh (dùng để chuyển đổi tọa độ chuẩn hóa sang pixel)

    Returns:
        _type_: _description_ trả về list các bounding box GT dưới dạng dict {"box": [x1, y1, x2, y2], "class": class_id}
    """
    boxes = []

    with open(label_path, "r") as f:
        for line in f:
            cls, x, y, bw, bh = map(float, line.strip().split())

            x1 = (x - bw/2) * w 
            y1 = (y - bh/2) * h 
            x2 = (x + bw/2) * w
            y2 = (y + bh/2) * h

            boxes.append({
                "box": [x1,y1,x2,y2],
                "class": int(cls)
            })

    return boxes

# Hàm lấy bounding box dự đoán của mô hình kèm theo class_id để tính IoU với Ground Truth
def get_pred_boxes_with_class(results):
    """_summary_ Hàm lấy bounding box dự đoán của mô hình kèm theo class_id để tính IoU với Ground Truth. 
    Định dạng trả về là list các dict {"box": [x1, y1, x2, y2], "class": class_id}

    Args:
        results (_type_): _description_ đầu vào là kết quả dự đoán của mô hình YOLOv11 sau khi chạy predict, 
        chứa thông tin về bounding box và class_id của các đối tượng được phát hiện.

    Returns:
        _type_: _description_ trả về list các dict {"box": [x1, y1, x2, y2], "class": class_id} đại diện cho các bounding box dự đoán của mô hình kèm theo class_id tương ứng.
    """
    boxes = []

    xyxy = results[0].boxes.xyxy.cpu().numpy()
    cls  = results[0].boxes.cls.cpu().numpy().astype(int)

    for b, c in zip(xyxy, cls):
        boxes.append({
            "box": b.tolist(),
            "class": int(c)
        })

    return boxes


def analyze_errors(gt_boxes, pred_boxes, iou_thresh=0.5):
    """_summary_ Hàm phân tích lỗi giữa Ground Truth và Prediction, 
    phân loại thành True Positive (TP), False Positive (FP), False Negative (FN), lỗi vị trí (Localization Errors), lỗi phân loại (Classification Errors) 
    và các trường hợp trùng lặp (Duplicates).

    Args:
        gt_boxes (_type_): _description_ list các bounding box Ground Truth dưới dạng dict {"box": [x1, y1, x2, y2], "class": class_id}
        pred_boxes (_type_): _description_ list các bounding box dự đoán của mô hình dưới dạng dict {"box": [x1, y1, x2, y2], "class": class_id}
        iou_thresh (float, optional): _description_. Defaults to 0.5. Ngưỡng IoU để xác định một dự đoán có khớp với Ground Truth hay không. 
        Nếu IoU >= iou_thresh, dự đoán được coi là khớp với GT. 
        Tuy nhiên, nếu IoU thấp hơn nhưng vẫn được coi là TP, sẽ được phân loại thành lỗi vị trí (Localization Error). 
        Nếu class_id không khớp dù IoU cao, sẽ được phân loại thành lỗi phân loại (Classification Error). 
        Các dự đoán trùng lặp (Duplicate) là những trường hợp có nhiều dự đoán khớp với cùng một GT.

    Returns:
        _type_: _description_ trả về 5 list: TP, FP, FN, loc_errors, cls_errors, duplicates
        - TP: list các dự đoán đúng (True Positive)
        - FP: list các dự đoán sai (False Positive)
        - FN: list các GT bị bỏ sót (False Negative)
        - loc_errors: list các dự đoán có lỗi vị trí (Localization Errors)
        - cls_errors: list các dự đoán có lỗi phân loại (Classification Errors)
        - duplicates: list các dự đoán trùng lặp (Duplicates)
    """
    TP, FP, FN = [], [], []
    loc_errors = []
    cls_errors = []
    duplicates = []

    matched_gt = set()

    # --- match GT vs Pred ---
    for i, pred in enumerate(pred_boxes):
        best_iou = 0
        best_j = -1

        for j, gt in enumerate(gt_boxes):
            iou = compute_iou_simple(pred["box"], gt["box"])
            if iou > best_iou:
                best_iou = iou
                best_j = j

        if best_iou >= iou_thresh:
            if best_j not in matched_gt:
                gt = gt_boxes[best_j]

                # classification error
                if pred["class"] != gt["class"]:
                    cls_errors.append(pred)
                else:
                    TP.append(pred)

                # localization error (IoU thấp nhưng vẫn match)
                if best_iou < 0.7:
                    loc_errors.append(pred)

                matched_gt.add(best_j)
            else:
                duplicates.append(pred)
        else:
            FP.append(pred)

    # --- FN ---
    for i, gt in enumerate(gt_boxes):
        if i not in matched_gt:
            FN.append(gt)

    return TP, FP, FN, loc_errors, cls_errors, duplicates


def draw_all_errors(img, TP, FP, FN, loc, cls, dup):
    """_summary_ Hàm vẽ bounding box của tất cả các loại lỗi (TP, FP, FN, Localization Errors, Classification Errors, Duplicates) lên ảnh để trực quan hóa kết quả phân tích lỗi.

    Args:
        img (_type_): _description_ ảnh gốc chưa được vẽ bounding box lỗi, định dạng là numpy array (BGR)
        TP (_type_): _description_ list các dự đoán đúng (True Positive), mỗi phần tử là dict {"box": [x1, y1, x2, y2], "class": class_id}
        FP (_type_): _description_ list các dự đoán sai (False Positive), mỗi phần tử là dict {"box": [x1, y1, x2, y2], "class": class_id}
        FN (_type_): _description_ list các Ground Truth bị bỏ sót (False Negative), mỗi phần tử là dict {"box": [x1, y1, x2, y2], "class": class_id}
        loc (_type_): _description_ list các dự đoán có lỗi vị trí (Localization Errors), mỗi phần tử là dict {"box": [x1, y1, x2, y2], "class": class_id}
        cls (_type_): _description_ list các dự đoán có lỗi phân loại (Classification Errors), mỗi phần tử là dict {"box": [x1, y1, x2, y2], "class": class_id}
        dup (_type_): _description_ list các dự đoán trùng lặp (Duplicates), mỗi phần tử là dict {"box": [x1, y1, x2, y2], "class": class_id}

    Returns:
        _type_: _description_ trả về ảnh đã được vẽ bounding box của tất cả các loại lỗi, định dạng là numpy array (BGR).
    """
    img = img.copy()

    # TP - xanh lá
    for obj in TP:
        x1,y1,x2,y2 = map(int, obj["box"])
        cv2.rectangle(img, (x1,y1), (x2,y2), (0,255,0), 2)

    # FP - đỏ
    for obj in FP:
        x1,y1,x2,y2 = map(int, obj["box"])
        cv2.rectangle(img, (x1,y1), (x2,y2), (0,0,255), 2)

    # FN - xanh dương
    for obj in FN:
        x1,y1,x2,y2 = map(int, obj["box"])
        cv2.rectangle(img, (x1,y1), (x2,y2), (255,0,0), 2)

    # Localization - vàng
    for obj in loc:
        x1,y1,x2,y2 = map(int, obj["box"])
        cv2.rectangle(img, (x1,y1), (x2,y2), (0,255,255), 2)

    # Classification - tím
    for obj in cls:
        x1,y1,x2,y2 = map(int, obj["box"])
        cv2.rectangle(img, (x1,y1), (x2,y2), (255,0,255), 2)

    # Duplicate - cam
    for obj in dup:
        x1,y1,x2,y2 = map(int, obj["box"])
        cv2.rectangle(img, (x1,y1), (x2,y2), (0,165,255), 2)

    return img



def save_full_summary_txt(save_path, gt_counts, pred_counts, TP, FP, FN, loc, cls, dup):
    """_summary_ Hàm lưu báo cáo tổng kết phân tích lỗi vào file txt, bao gồm số lượng từng class trong Ground Truth và Prediction,
    và thống kê số lượng lỗi của từng loại (TP, FP, FN, Localization Errors, Classification Errors, Duplicates).

    """
    gt_total = sum(gt_counts.values())
    pred_total = sum(pred_counts.values())

    content = f"""Ground Truth:
    Total: {gt_total}
    {format_counts(gt_counts)}

    Prediction:
    Total: {pred_total}
    {format_counts(pred_counts)}
    
    Error Analysis:
    TP: {len(TP)}
    FP: {len(FP)}
    FN: {len(FN)}
    Localization errors: {len(loc)}
    Classification errors: {len(cls)}
    Duplicate detections: {len(dup)}
    """

    with open(save_path, "w", encoding="utf-8") as f:
        f.write(content)
        
 
```

```python
IMAGES_DIR = 'output/yolo_split/test/images'
LABELS_DIR = 'output/yolo_split/test/labels'
MODEL_PATH = r'YOLOv11_training\model_tuning\weights\best.pt' 
OUTPUT_VIS_DIR = 'result_vis'
CLASS_NAMES = ['car', 'cycle', 'bus', 'truck', 'van']
image_paths = glob.glob(os.path.join(IMAGES_DIR, '*.jpg')) 
density_threshold = 10 # Ngưỡng để xác định "mật độ cao"
saved_count = 0

# Lặp qua tất cả ảnh trong thư mục validation
for img_path in image_paths:
    img_name = os.path.basename(img_path) # Lấy tên file ảnh để tìm file nhãn tương ứng
    label_path = os.path.join(LABELS_DIR, img_name.replace('.jpg', '.txt')) 
    
    # Chạy dự đoán
    results = model.predict(source=img_path, conf=0.25, verbose=False)
    h, w, _ = cv2.imread(img_path).shape # Lấy kích thước ảnh để chuyển đổi tọa độ GT từ chuẩn hóa sang pixel

    gt_boxes = get_gt_boxes_with_class(label_path, w, h) # Lấy bounding box GT kèm theo class_id để phân tích lỗi
    pred_boxes = get_pred_boxes_with_class(results) # Lấy bounding box dự đoán của mô hình kèm theo class_id để phân tích lỗi

    TP, FP, FN, loc, cls, dup = analyze_errors(gt_boxes, pred_boxes) # Phân tích lỗi giữa GT và Prediction
    pred_counts = get_pred_class_counts(results) # Đếm số lượng từng class trong kết quả dự đoán của mô hình
    num_vehicles = len(results[0].boxes) # Đếm số lượng xe trong ảnh dựa trên kết quả dự đoán của mô hình
    
    # Chỉ xử lý nếu ảnh có nhiều xe (mật độ cao)
    if num_vehicles >= density_threshold:
        # Lấy ảnh Ground Truth
        img_gt, class_counts = draw_ground_truth(img_path, label_path) # Vẽ bounding box GT lên ảnh và đếm số lượng từng class trong GT
        img_gt_rgb = cv2.cvtColor(img_gt, cv2.COLOR_BGR2RGB) # Chuyển đổi sang RGB để hiển thị bằng matplotlib
        
        # Lấy ảnh Prediction từ mô hình YOLO
        img_pred = results[0].plot() # Vẽ bounding box dự đoán của mô hình lên ảnh
        img = cv2.imread(img_path) # Đọc ảnh gốc để vẽ lỗi lên đó
        img_error = draw_all_errors(img, TP, FP, FN, loc, cls, dup) # Vẽ tất cả lỗi lên ảnh gốc để phân tích
        img_error_rgb = cv2.cvtColor(img_error, cv2.COLOR_BGR2RGB) # Chuyển đổi sang RGB để hiển thị bằng matplotlib
        img_pred_rgb = cv2.cvtColor(img_pred, cv2.COLOR_BGR2RGB) # Chuyển đổi sang RGB để hiển thị bằng matplotlib
        
        # Vẽ 2 ảnh cạnh nhau để so sánh GT vs Prediction và ảnh lỗi
        fig, axes = plt.subplots(1, 3, figsize=(25, 10))

        axes[0].imshow(img_gt_rgb)
        axes[0].set_title("Ground Truth")
        axes[0].axis('off')

        axes[1].imshow(img_pred_rgb)
        axes[1].set_title("Prediction")
        axes[1].axis('off')

        axes[2].imshow(img_error_rgb)
        axes[2].set_title("Error Analysis")
        axes[2].axis('off')
        
        plt.tight_layout()
        
        # Lưu vào thư mục result_vis
        save_path = os.path.join(OUTPUT_VIS_DIR, f"compare_{img_name}")
        plt.savefig(save_path, dpi=150, bbox_inches='tight')
        plt.close() # Đóng plot để tránh tràn bộ nhớ
        txt_path = save_path.replace(".jpg", ".txt").replace(".png", ".txt")
        save_full_summary_txt(txt_path, class_counts, pred_counts, TP, FP, FN, loc, cls, dup)
        saved_count += 1
        # Lưu 5 ảnh
        if saved_count >= 5: 
            break

print(f"Đã lưu {saved_count} ảnh so sánh vào thư mục '{OUTPUT_VIS_DIR}'.")
```
